```python
#load results table
res_genes = pd.read_csv("gene_resAnno_only_male_231220.csv", 
                        usecols = ['Unnamed: 0','GENENAME', "log2FoldChange", "padj", "pvalue"])  
res_genes.rename(columns={"Unnamed: 0": "GENEID"}, inplace=True)
```

```python
res_genes
```

```python
#thresholds for DE genes
th_padj = 0.05 
th_log2FC = 1
```

```python
#gene lists
gene_list =  ["Cacna1a",
                    "Cacna1b",
                    "Cacna1c",
                    "Cacna1d",
                    "Cacna1e",
                    "Cacna1f",
                    "Cacna1g",
                    "Cacna1h",
                    "Cacna1i",
                    "Cacna2d1",
                    "Cacna2d2",
                    "Cacnb1",
                    "Cacnb2",
                    "Cacnb3",
                    "Cacnb4"
                    ]



                    
goi = res_genes[res_genes["GENENAME"].isin(gene_list)]
```

```python
goi.sort_values(by=["GENENAME"], ascending=True)
```

```python
# log2 transform
counts_genes_norm_log2 = pd.read_csv("results_gene_counts_normalized_only_male_231212.csv")  
counts_genes_norm_log2.rename(columns={"Unnamed: 0": "GENEID"}, inplace = True)
cols = counts_genes_norm_log2.columns
counts_genes_norm_log2[cols[1:]] = counts_genes_norm_log2[cols[1:]].apply(np.log2, errors='coerce')
```

```python
#merge counts and results

######################################
goi_counts_norm_log2 = goi.merge(counts_genes_norm_log2, how = "left", 
                           left_on='GENEID', 
                           right_on='GENEID', 
                           validate="one_to_one")
goi_counts_norm_log2.set_index('GENENAME', inplace=True)
######################################
```

```python
goi_counts_norm_log2.sort_values(by=["GENENAME"], ascending=True)
```

```python
goi_counts_norm_log2\
.sort_values(by=["GENENAME"], ascending=True)\
.to_excel("data_l2fc_p_log2counts_calcium_channels_only_male_231212.xlsx")
```
