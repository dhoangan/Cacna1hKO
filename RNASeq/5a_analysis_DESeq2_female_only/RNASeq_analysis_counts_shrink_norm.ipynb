```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#load count matrix (normalized)
counts_genes_norm = pd.read_csv("results_gene_counts_normalized_only_female_231212.csv")  
counts_genes_norm.rename(columns={"Unnamed: 0": "GENEID"}, inplace = True)
#rearrange columns
counts_genes_norm = counts_genes_norm[['GENEID', 
              'filtered_S1_quant_filtered',
              'filtered_S2_quant_filtered',
              'filtered_S4_quant_filtered',
              'filtered_S9_quant_filtered',
              'filtered_S11_quant_filtered',
              'filtered_S12_quant_filtered',]]
```

```python
counts_genes_norm
```

```python
#load results table
res_genes = pd.read_csv("gene_resAnno_only_female.csv", 
                        usecols = ['Unnamed: 0','GENENAME', "log2FoldChange", "padj", "pvalue"])  
res_genes.rename(columns={"Unnamed: 0": "GENEID"}, inplace=True)
```

```python
res_genes
```

```python
#gene lists
gene_list =  ["Cacna1a",
                    "Cacna1b",
                    "Cacna1c",
                    "Cacna1d",
                    "Cacna1e",
                    "Cacna1f",
                    "Cacna1g",
                    "Cacna1h",
                    "Cacna1i",
                    "Cacna2d1",
                    "Cacna2d2",
                    "Cacnb1",
                    "Cacnb2",
                    "Cacnb3",
                    "Cacnb4"
                    ]

goi = res_genes[res_genes["GENENAME"].isin(gene_list)]
```

```python
goi.sort_values(by=["GENENAME"], ascending=True)
```

```python
# log2 transform
counts_genes_norm_log2 = pd.read_csv("results_gene_counts_normalized_only_female_231212.csv")   
counts_genes_norm_log2.rename(columns={"Unnamed: 0": "GENEID"}, inplace = True)
cols = counts_genes_norm_log2.columns
counts_genes_norm_log2[cols[1:]] = counts_genes_norm_log2[cols[1:]].apply(np.log2, errors='coerce')
```

```python
#merge log2 counts and LFC padj results

######################################
goi_counts_norm_log2 = goi.merge(counts_genes_norm_log2, how = "left", 
                           left_on='GENEID', 
                           right_on='GENEID', 
                           validate="one_to_one")
goi_counts_norm_log2.set_index('GENENAME', inplace=True)
######################################
```

```python
goi_counts_norm_log2.\
sort_values(by=["GENENAME"], ascending=True).\
to_excel("data_l2fc_p__log2counts_calcium_channels_only_female_231212.xlsx")
```
