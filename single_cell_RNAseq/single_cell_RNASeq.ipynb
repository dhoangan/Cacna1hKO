```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

color_palette = sns.color_palette("colorblind")
sns.set_theme(palette=color_palette, font="Arial", font_scale=1.0, style="white")
plt.style.use('paper.mplstyle')
```

```python
data = pd.read_csv("calcium_gene_counts_in_ZG_normalized.csv")
```

```python
data.head()
```

```python
fig, g = plt.subplots(figsize=(2, 2))


g = sns.kdeplot(data=data, x="Cacna1c", y="Cacna1d", fill=True)
g = sns.scatterplot(data=data, x="Cacna1c", y="Cacna1d", color="black", marker=".", alpha=0.2)

g.set_ylabel("Cacna1d expression")
g.set_xlabel("Cacna1c expression")

sns.despine()
```

```python
fig, g = plt.subplots(figsize=(2, 2))

g = sns.kdeplot(data=data, x="Cacna1h", y="Cacna1c", fill=True)
g = sns.scatterplot(data=data, x="Cacna1h", y="Cacna1c", color="black", marker=".", alpha=0.1)

g.set_ylabel("Cacna1c expression")
g.set_xlabel("Cacna1h expression")

sns.despine()
```
