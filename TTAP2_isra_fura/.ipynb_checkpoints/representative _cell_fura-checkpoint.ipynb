```python
import calim
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.patches as mpatches
import seaborn as sns
import pickle
import pandas as pd
import numpy as np

color_palette = sns.color_palette("colorblind")
sns.set_theme(palette=color_palette, font="Arial", font_scale=.7, style="white")
```

```python
#Stand Kalibration 10.03.2020 f√ºr das Scientifica-Setup
# mit Korrektur vom 19.05.2021 (ab hier zu verwenden)
kd = 417.1324107961717
R_max = 1.9267297757725788*1.305670312829699
R_min = 0.27585714285714286*0.8222568831620652
sf2_by_sb2 = 2273/413



def calcium(data, kd, R_max, R_min, sf2_by_sb2):
    return kd*((data-R_min)/(R_max-data))*sf2_by_sb2
```

```python
# Names of used data files


filename_wt =  "doppel_perf_fura_data+ctrl_burst_selection_230601.hdf"

data_wt_hdf = calim.Project()
data_wt_hdf.from_hdf(filename_wt)
```

```python
start_frame_wt = 0
duration = 22400

wt_recording="230224_slice04_ttap2_isra_results.csv"
wt_cell = "Mean2"
dt = data_wt_hdf.recordings[wt_recording].dt

wt_data = pd.DataFrame(data_wt_hdf.recordings[wt_recording].cells[wt_cell]\
                       .raw_data[start_frame_wt:start_frame_wt+duration])
wt_data.columns = ["Fluorescence (a.u)"]
wt_data["mean_ca"] = calcium(wt_data["Fluorescence (a.u)"], kd, R_max, R_min, sf2_by_sb2)
wt_data["Time (s)"] = np.arange(0, duration*dt, dt)

g = sns.relplot(data = wt_data[2025:], y="mean_ca",x="Time (s)",
            kind="line", height=2, aspect=3, color=color_palette[0], lw=0.5)

plt.ylim(505, 1850)

plt.hlines(y=530, xmin=2020, xmax=2200, linewidth=2, color='k')
plt.text(2035, 570, "3 min", size=8)



# Create a Rectangle patch isra/ttap2

g.ax.add_patch(plt.Rectangle((445, 1785),1200, 100, facecolor='w', linewidth=1, edgecolor = "black", 
                              clip_on=False, zorder=20))
plt.text(460, 1800, "100 pM Ang II", ha="left", color="black", zorder=25, size=8) 

g.ax.add_patch(plt.Rectangle((1645, 1785),600, 100, facecolor='w', linewidth=1, edgecolor = "black", 
                              clip_on=False, zorder=20))
plt.text(1665, 1800, "500 pM", ha="left", color="black", zorder=25, size=8) 

g.ax.add_patch(plt.Rectangle((1045, 1660),1200, 100, facecolor='w', linewidth=1, edgecolor = "black", 
                              clip_on=False, zorder=20))
plt.text(1065, 1675, "TTA-P2 + Isradipine", ha="left", color="black", zorder=25, size=8)



g.set(xticklabels=[])
g.set(xlabel=None)
plt.tick_params(bottom=False)  # remove the ticks

g.set(ylabel="Mean [Ca$^{2+}$]$_i$ (nmol/l)")


#g.set(yticks=[600, 700, 800, 900, 1000 ,1100, 1200])
#plt.tick_params(left=False)  # remove the ticks

c = data_wt_hdf.recordings[wt_recording].cells[wt_cell]
burst_list = [x.frame/10 for x in c.baseline if x.use] # Get list of burst frames
plt.plot(burst_list, [710]*len(burst_list),linestyle='',marker='s', ms=1, color="k")
sns.despine(bottom=True)
```
